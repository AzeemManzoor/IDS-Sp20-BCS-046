```python
import pandas as pd
import numpy as np
```

```python
#Load the dataset (csv file) into a Pandas DataFrame
df = pd.read_csv('the-hello-dataset-fa22.csv')
df['index'] = df.index
df

##Data Normalization Steps
# """
# 1. First Letter of each data(stringtype) set as capital.
# 2. Correct the spellings of Male and Female.
# 3. Convert percentage into simple form.
# 4. Add a new column named Index
# """
```

```python
#Print the list of all students whose first name starts with letter the 'H'.
df1=df['Name']
for h in df1:
    if h[0]=='H':
        print(h)
```

```python
# Print the total number of students who have a three words name (first-middle-surname).
c=0
df1=df['Name'].str.split()
for i in df1:
   
    if len(i)==3:
        c+=1
        print(i)
print("\nTotal number of students who have three words name (first-middle-surname).= ",c)
        
        
#     print(i[0:1])               #To print first 2 names  
#     print(i[0:2])               # To print first 3 names
```

```python
#Plot a pie chart to show the ratio of male and female students.
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
newdf = df['Gender'].value_counts()
newdf
```

```python
newdf = df['Gender'].value_counts().rename_axis('gender_types').reset_index(name='total')
print(newdf)
total = newdf.total.sum()
print("\nTotal of males and females: ",total)
```

```python
my_labels = newdf.gender_types
my_values = newdf.total
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')

ax.pie(my_values, labels = my_labels, autopct='%1.2f%%')            #This will find auto_percentage
plt.show()
```

```python
# Plot the HSSC-1 marks of all male vs female students on a scatter plot.
males=df[['HSSC-1']]
Index=df[['HSSC-1']]
x1=Index[df['Gender'] == 'Male']
x2=Index[df['Gender'] == 'Female']

m=males[df['Gender'] == 'Male']
f=males[df['Gender'] == 'Female']

m
```

```python
plt.scatter(x1,m,label='HSSC-1 Marks of Male Students ')
plt.scatter(x2,f,label='HSSC-1 Marks of Fe-male Students')
plt.legend()
```

```python
#Plot the favorite colors of male vs female students on a bar chart.
males=df['index']
Index=df['FavoriteColor']
b1=Index[df['Gender'] == 'Male']
b2=Index[df['Gender'] == 'Female']

m=males[df['Gender'] == 'Male']
f=males[df['Gender'] == 'Female']

b1
```
