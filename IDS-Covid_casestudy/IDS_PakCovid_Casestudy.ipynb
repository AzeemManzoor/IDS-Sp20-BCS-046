```python
#create a map on Pakistan location
pakistan_map = folium.Map(location = [30.3753,69.3451],zoom_start=4.5,tiles='OpenStreetMap') 
```

```python
#load the dataset (csv file) into the Pandas dataframe
df = pd.read_csv("CoronaProvincialDataPakistan.csv")

# df

df = df.drop('SNo', axis=1)
aggregation_functions = {'Province':'first','New Cases': 'sum','New Deaths': 'sum', 'New Recovered': 'sum', 'Date': 'max'}
df_new1 = df.groupby(df['Province']).aggregate(aggregation_functions)
df_new = df.groupby('Province', as_index=False).aggregate(aggregation_functions).reindex(columns=df.columns)
val1 = 1
df_new.insert(loc=5, column="Latitude", value=[33.9259,28.4907,35.8026,33.7205,34.9526,32.6675,28.964802,31.1704,25.8943])
df_new.insert(loc=6, column="Longitude", value=[73.7810,65.0958,74.9832,73.0405,72.3311,69.8597,61.585609,72.7097,68.5247])
# df_new["Longitude"] = 2
# df_new = df.groupby(['Province','New Deaths'])

df_new
```

```python

#calculate total number of cases using a loop
t = 0
for c in df_new['New Cases']:
  t = int(c)+int(t)
t
```

```python
# Latitude=30.3753
# Longitude=69.3451

#mark circles on the map and display the map
for prov,tc,de,re,lat,long in zip(list(df_new['Province']),list(df_new['New Cases']),list(df_new['New Deaths']),list(df_new['New Recovered']),list(df_new['Latitude']),list(df_new['Longitude'])):
    
    folium.CircleMarker(radius = (tc/t)*100,
        location = [lat,long],
                       color = 'red',
                       fill = False,
                       fill_color = 'blue').add_to(pakistan_map)
    xyz = '<strong>Province:  ' + prov + '</strong><br>' + 'Total Cases: ' + str(tc) + '<br>' + 'Total Deaths: '+ str(de) + '<br>' + 'Total Recoveries: ' + str(re) + '<br>'+'<br>' + '<strong>This data is from 23-March to 23 March 2020  ' 
    iframe = folium.IFrame(xyz, width=200, height=200)
    popupData = folium.Popup(iframe, max_width=500)
    folium.Marker(location = [lat,long], popup=popupData).add_to(pakistan_map)
    
pakistan_map
```
